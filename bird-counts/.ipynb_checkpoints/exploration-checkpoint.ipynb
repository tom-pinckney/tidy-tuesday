```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bird_counts = pd.read_csv('data/bird_counts.csv')

bird_counts.head(10)
```

```python
# Total counts per year
yearly_counts = bird_counts.groupby(['year', 'total_hours']).sum()[['how_many_counted']].reset_index()
yearly_counts['ave_hour'] = yearly_counts['how_many_counted'] / yearly_counts['total_hours']

# make long
yearly_counts = pd.melt(yearly_counts, id_vars= ['year'], value_vars = ['how_many_counted','total_hours', 'ave_hour'])

sns.set(palette = 'viridis')
g = sns.FacetGrid(yearly_counts, col = 'variable', height = 6, sharey= False)
g = g.map(sns.lineplot, 'year', 'value')
g
```

## Birds counted every year
After 1940

```python
# How many years is a species not observed
bird_counts_post_1940 = bird_counts[bird_counts.year > 1980]

bird_counts_post_1940['has_zero'] = bird_counts_post_1940['how_many_counted'] == 0
number_zero_years = bird_counts_post_1940.groupby('species').sum()['has_zero']

# Which birds were observed every year
birds_no_zeros = number_zero_years[number_zero_years == 0].index

# Data on birds observed every year
every_year_birds = bird_counts_post_1940[bird_counts_post_1940['species'].isin(birds_no_zeros)]

# Data for labels
label = every_year_birds.loc[(every_year_birds.species == "European Starling") & (every_year_birds.year == 1980)]

f, ax = plt.subplots(figsize = (12, 9))
common_birds = sns.lineplot(x = 'year',
                            y = 'how_many_counted',
                            hue = 'species',
                            legend = False,
                            palette = 'viridis',
                            data = every_year_birds)

for bird in range(0, label.shape[0]):
    common_birds.text(label.year.values[bird] + 0.2,
                      label.how_many_counted.values[bird],
                      label.species.values[bird])

# for line in range(0,df.shape[0]):
#     p1.text(df.x[line]+0.2, df.y[line], df.group[line], horizontalalignment='left', size='medium', color='black', weight='semibold'
```

## What birds are increasing, decreasing, or staying the same

```python
# Using linear regression
import statsmodels.api as sm

starling = bird_counts_post_1940[bird_counts_post_1940.species == "European Starling"].reset_index()

# Try on one species
y = starling.how_many_counted.values
x = starling.index.values
x = sm.add_constant(x)
mod = sm.OLS(y, x).fit()
params = mod.params


mod.params
```

```python
# Define a function to apply
def group_ols(data):
    data.reset_index()
    y = data.how_many_counted_by_hour.values
    x = data.index.values
    x = sm.add_constant(x)
    mod = sm.OLS(y, x).fit()
    params = mod.params
    
    return np.append(params, mod.mse_resid)

group_ols(starling)
```

```python
# Function works for one group, lets try applying it 
# Remove all years with NaN for house
mod_data = bird_counts_post_1940[~bird_counts_post_1940.total_hours.isnull()]


bird_coefs = mod_data.groupby('species').apply(group_ols).reset_index()
bird_coefs.columns = ['species', 'coef']

# Split parameters into intercept and coeficient columns
bird_coefs[['intercept', 'coef', 'mse_resid']] = pd.DataFrame(bird_coefs.coef.values.tolist(), index = bird_coefs.index)

bird_coefs.head()
```

```python
# Birds that increse the most
top_5_birds = bird_coefs.sort_values('coef', ascending = False).head().species.values[0:6]

# Data on birds observed every year
increasing_birds = bird_counts_post_1940[bird_counts_post_1940['species'].isin(top_5_birds)]

f, ax = plt.subplots(figsize = (12, 9))
common_birds = sns.lineplot(x = 'year',
                            y = 'how_many_counted',
                            hue = 'species',
                            palette = 'viridis',
                            data = increasing_birds)


plt.suptitle("Bird Species increasing the most since 1940", 
             x = 0.33,
             y=0.932, 
             fontsize=18)
plt.title("Estimated using Linear Regression", fontsize=14, loc = 'left')
```

```python
# Birds that decrease
bottom_5_birds = bird_coefs.sort_values('coef', ascending = True).head().species.values

# Data on birds observed every year
decreasing_birds = bird_counts_post_1940[bird_counts_post_1940['species'].isin(bottom_5_birds)]

f, ax = plt.subplots(figsize = (12, 9))
common_birds = sns.lineplot(x = 'year',
                            y = 'how_many_counted',
                            hue = 'species',
                            palette = 'viridis',
                            data = decreasing_birds)

plt.suptitle("Bird Species decreasing the most since 1940", 
             x = 0.334,
             y=0.932, 
             fontsize=18)
plt.title("Estimated using Linear Regression", fontsize=14, loc = 'left')
```

```python
# Birds with most non-linear patterns
non_linear_5_birds = bird_coefs.sort_values('mse_resid', ascending = False).head().species.values

# Data on birds observed every year
non_linear_birds = bird_counts_post_1940[bird_counts_post_1940['species'].isin(non_linear_5_birds)]

f, ax = plt.subplots(figsize = (12, 9))
common_birds = sns.lineplot(x = 'year',
                            y = 'how_many_counted',
                            hue = 'species',
                            palette = 'viridis',
                            data = non_linear_birds)

plt.suptitle("Bird Species with non-linear patterns", 
             x = 0.295,
             y=0.932, 
             fontsize=18)
plt.title("Estimated using Linear Regression residuals", fontsize=14, loc = 'left')
```
