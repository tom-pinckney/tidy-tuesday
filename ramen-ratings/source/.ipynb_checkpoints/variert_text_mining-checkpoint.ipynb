# Mining Varieties
Words, bigrams, and trigrams of variety

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import re
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Set working directory to highest level of repo
os.chdir("..")
print(os.getcwd())
```

```python
# Read in and view data
ratings = pd.read_csv('data/ramen-ratings.csv')

ratings['variety_clean'] = ratings.variety.str.replace('[^a-zA-Z0-9]', ' ')

ratings.tail()

# Remove rows without ratings
ratings_no_nan = ratings.loc[~ratings.stars.isnull(), :]
```

## Finding words leading to higher star ratings
We will create a sparse matrix and then use lasso regression to pick words that have a large effect on ratings. 

```python
# For creating sparse matrix
countvec = CountVectorizer()

# Convert matrix to dataframe
words = pd.DataFrame(countvec.fit_transform(ratings_no_nan.variety_clean).toarray(), columns = countvec.get_feature_names())

words.shape
```

```python
from sklearn.linear_model import LassoCV

lasso_reg = LassoCV(cv = 5)

lasso_reg.fit(words, ratings_no_nan['stars'].values)
```

```python
lasso_reg.score(words, ratings_no_nan['stars'].values)
```

```python
words.columns[lasso_reg.coef_ < 0]
```

```python
words.columns[lasso_reg.coef_ > 0]
```

Plot top ten positive words and top ten negative words


```python
# Create a dataframe of just the words and coefficients
word_coefs = pd.DataFrame({'word': words.columns, 
                           'coef' : lasso_reg.coef_,
                           'abs_coef' : np.abs(lasso_reg.coef_)})

# Select the top 20 largest coefs
top_20_words = word_coefs.sort_values('abs_coef', ascending = False).iloc[0:20, :]

# Plot results
sns.set(context = 'talk',
        style = 'whitegrid')
f, ax = plt.subplots(figsize=(9, 6))
sns.barplot(y = 'word',
            x = 'coef',
            order = top_20_words.sort_values('coef', ascending = False)['word'].values,
            data = top_20_words)
ax.set(ylabel = '')
sns.despine(left=True, bottom=True)
```
